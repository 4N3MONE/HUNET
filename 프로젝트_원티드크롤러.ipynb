```python
from bs4 import BeautifulSoup 
from selenium import webdriver
import datetime
import time
import sys
import pandas as pd
```

```python
#사용자로부터 어느 직군을 크롤링해야할지 입력받는다

jobgroup = input("""어느 직군의 채용공고를 크롤링 하시겠습니까?
1. 개발
2. 경영/비즈니스
3. 디자인
4. 미디어
5. 게임 제작
6. 인사
7. 물류/무역
8. 금융""")
if jobgroup == "1" :  
    jobquery = "518"
    csvname = '개발'
if jobgroup == "2" :
    jobquery = "507"
    csvname = '경영&비즈니스'
if jobgroup == "3" :
    jobquery = "511"
    csvname = '디자인'
if jobgroup == "4" :
    jobquery = "524"
    csvname = '미디어'
if jobgroup == "5" :
    jobquery = "959"
    csvname = '게임 제작'
if jobgroup == "6" :
    jobquery = "517"
    csvname = '인사'
if jobgroup == "7" :
    jobquery = "532"
    csvname = '물류무역'
if jobgroup == "8" :
    jobquery = "508"
    csvname = '금융'


filename = 'c:\\tmp\\크롤링\\원티드_크롤링_'+csvname+'.csv'
```

```python
#원티드 사이트 접속
path = "c:\\tmp\\chromedriver.exe"
driver = webdriver.Chrome(path)
driver.get("https://www.wanted.co.kr/wdlist/{}?country=kr&job_sort=job.popularity_order&years=-1&locations=all" .format(jobquery))
#스크롤 내리기

x = 0
for x in range(0,10) :
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    time.sleep(5)
    x +=1

#공고들로부터 url 뽑아오기
full_html = driver.page_source
soup = BeautifulSoup(full_html, 'html.parser')
content_list = soup.find_all('div', class_="_3D4OeuZHyGXN7wwibRM5BJ")
url_list = []
for content in content_list:
    url = content.find('a')['href']
    url_list.append(url)
# print(url_list)
no=1
dict1 = []
#url 리스트에서 순차적으로 접속해서 크롤링
for i in range(len(url_list)) :

    target_info = {}
      
    url = "'"+url_list[i]+"'"
    driver.switch_to_window(driver.window_handles[0])
    driver.execute_script('window.open({});'.format(url))
    time.sleep(2)

    driver.switch_to_window(driver.window_handles[1])
    html2 = driver.page_source
    soup2 = BeautifulSoup(html2, 'html.parser')

    # company_name = soup2.find('a')['data-company-name']
    
    contents = soup2.select('section._1LnfhLPc7hiSZaaXxRv11H p')
    content_list = []
    for content in contents:
        content_list.append(content.text)
    content_str = ' '.join(content_list)
    # content_str2 = content_str.split('주요업무')[1]
    # content_str3 = content_str.split('복지')[0]
    target_info['content'] = content_str
    dict1.append(target_info)
    no += 1
    driver.close()
    print("번호 : {}" .format(no))
    print('내용 : {}' .format(content_str))
    contents = None
    content_str = None
```

```python
#데이터 저장
result_df = pd.DataFrame(dict1)
result_df.to_csv(filename, encoding="utf-8-sig" , index=False )
result_df = pd.DataFrame(dict1)
```
